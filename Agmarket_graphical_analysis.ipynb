```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=  pd.read_csv("/content/agmarket file.csv")
data.head()
```

```python
data.tail()
```

```python
data.info()
```

```python
data['Price Date'] = pd.to_datetime(data['Price Date'],format = '%d-%b-%y')
data = data.sort_values(by = 'Price Date' ).reset_index().drop(['index','Sl no.'], axis = 1)

data.head(2).append(data.tail(2))
```

```python
import plotly.express as px  

market = ['Samsabad', 'Achnera', 'Jagnair', 'Khairagarh', 'Fatehpur Sikri','Jarar', 'Agra', 'Fatehabad']

for val in market:
  
  temp =  data[data['Market Name'] == val]
  

  fig = px.line( temp, x = 'Price Date' ,y = 'Modal Price (Rs./Quintal)', title = 'Price pattern of ' +val +" market" ,width=1200, height=400)
  fig.add_scatter(x=temp['Price Date'], y=temp['Min Price (Rs./Quintal)'], mode='lines' ,name = 'Minimum price')
  fig.add_scatter(x=temp['Price Date'], y=temp['Max Price (Rs./Quintal)'], mode='lines',name = 'Maximum price')

  fig.update_xaxes(rangeslider_visible=True)
  fig.show()
```
